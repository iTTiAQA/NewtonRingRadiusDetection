```python
# Download data from open datasets
import zipfile
  
# 假设你的ZIP文件在OBS中，你需要先将它下载到Notebook的文件系统中
# 这里只是演示如何解压已经在本地的ZIP文件

# 指定ZIP文件的路径
zip_file_path = './Ideal.zip'

# 指定解压后的文件要存放的目录
extract_dir = './'

# 使用zipfile库解压ZIP文件
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)
```
